```julia
import random

# Function to generate a 4-digit OTP
def generate_otp():
    otp = random.randint(100000, 999999)  # Random integer between 1000 and 9999
    return otp

# Print the OTP
otp = generate_otp()
print(f"Your 4-digit OTP is: {otp}")

```

```julia
# Function to calculate the average of numbers
def calculate_average(*numbers):
    if len(numbers) == 0:
        return "No numbers provided"
    return sum(numbers) / len(numbers)

# Example usage: Pass numbers as arguments
numbers = [8888888, 8888888888888, 76543379, 12345678, 5567890987654321234567]  # You can replace this list with any numbers
average = calculate_average(*numbers)

print(f"The average is: {average}")
```
